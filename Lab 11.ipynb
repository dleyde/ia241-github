```python
from urllib import request
import json
from pprint import pprint

census_api_key = '' # get your key from https://api.census.gov/data/key_signup.html
url1 = f"https://api.census.gov/data/2016/acs/acs5?get=B01001_001E,NAME&for=county:*&in=state:51&key={census_api_key}"
url2 = f"https://api.census.gov/data/2016/acs/acs5?get=B01001_002E,NAME&for=county:*&in=state:51&key={census_api_key}"

response1 = requests.get(url1)
response2 = requests.get(url2)

data1 = response1.json()[1:]
data2 = response2.json()[1:]

county_data = {}
for row in data1:
    total_pop = int(row[0])
    county_name = row[1]
    county_data[county_name] = {"total_pop": total_pop}

for row in data2:
    male_pop = int(row[0])
    county_name = row[1]
    county_data[county_name]["male_pop"] = male_pop

for county in county_data:
    total_pop = county_data[county]["total_pop"]
    male_pop = county_data[county]["male_pop"]
    ratio = male_pop / total_pop
    county_data[county]["male_to_total_ratio"] = ratio

sorted_counties = sorted(county_data.items(), reverse=True, key=lambda x: x[1]["male_to_total_ratio"])

print(sorted_counties[0])
```
