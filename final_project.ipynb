Covid-19 is a respiratory virus caused by the SARS-CoV-2 virus. It began in 2019 affecting the world population. This disease is highly contagious, however with the development of vaccinations, its spread has slowed.

```python
df = pd.read_excel('s3://dleyde-ia240-s3/covid_data.xls')
```

## Estonia Data

```python
estonia_data = df.loc[ df['countriesAndTerritories']=='Estonia']
estonia_data[:10]
```

## Number of Cases related to Number of Deaths

```python
estonia_data.plot.scatter(x='cases',y='deaths',c= 'month')
```

It illustrates the number of deaths as it relates to the number of cases in Estonia. Darker colors indicates the end of the year while the ligher colors indicate the beginning of the year. Most of the cases were within the first 3 months of Covid-19 present in Estonia. Due to the small number of deaths in Estonia from Covid-19, this graph appears to have multiple gaps.

## Sum of Covid Cases per Month in Estonia

```python
sum_cases_per_month = estonia_data.groupby('dateRep').sum()['cases']
sum_cases_per_month.plot()
```

This graph illustrates the number of Covid cases per month. The cases generally stayed below 600 per month, with the largest spike peaking at around 750 in the month of December. There are large spikes shown because that is when the Covid data was reported for that month.

## Covid-19 Trends Over Time

```python
sum_cases_per_day=estonia_data.groupby('month').sum()['cases']
sum_cases_per_day.plot()
```

The graph shows a trend that the number of Covid cases had increased throughout the year of 2020. There was a sharp increase in the month of October as winter began, and peaked right before December at 7,500. At the end of the year the number of cases had slightly dropped.

## Conclusion & Limitations

We used python code to generate graphs to determine the relation between Covid cases per month and throughout the year, as well as the relation between Covid cases and deaths in Estonia. The primary concluson that we drew based on the graphs is that Covid-19 cases had steadily increased throughout the year the 2020, and peaked in December. The main limitaiton for the visualizations is the lack of Covid data in Estonia. This is due to the small population, since Estonia only has a little over 1 million people. Another limitation that we had was access to only one source of data, that being the European Center for Disease Prevention and Control. In summary, with these graphs we can predict that Covid-19 cases would decrease in Estonia due to the prevalence of vaccinations, increased PPE, and safety guidelines.
